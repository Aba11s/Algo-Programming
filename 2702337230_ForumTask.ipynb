```python
# Forum
#Task 1

lloyd = {
 "name": "Lloyd",
 "homework": [90.0,97.0,75.0,92.0],
 "quizzes": [88.0,40.0,94.0],
 "tests": [75.0,90.0]
}
alice = {
 "name": "Alice",
 "homework": [100.0, 92.0, 98.0, 100.0],
 "quizzes": [82.0, 83.0, 91.0],
 "tests": [89.0, 97.0]
}
tyler = {
 "name": "Tyler",
 "homework": [0.0, 87.0, 75.0, 22.0],
 "quizzes": [0.0, 75.0, 78.0],
 "tests": [100.0, 100.0]
}
students = [lloyd, alice, tyler]

for student in students:
    print(student["name"])
    print(student["homework"])
    print(student["quizzes"])
    print(student["tests"])
    
def average(numbers):
    total = float(sum(numbers))
    return total/len(numbers)

def get_average(student):
    homework = average(student["homework"])
    quizzes = average(student["quizzes"])
    tests = average(student["tests"])
    return homework*0.1 + quizzes*0.3 + tests*0.6

def get_letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"
    
print(get_letter_grade(get_average(lloyd)))

def get_class_average(students):
    results = []
    for student in students:
        results.append(get_average(student))
    return average(results)

print(f"Class average: {get_class_average(students)}")
print(f"Class grade: {get_letter_grade(get_class_average(students))}")

#Forum task 2
courseNumber = ("COP 2510", "EGN 3000L", "MAC 2281", "MUH 3016", "PHY 2048")

course_names = {
    courseNumber[0]: "Programming Concepts",
    courseNumber[1]: "Foundations of Engineering Lab",
    courseNumber[2]: "Calculus I",
    courseNumber[3]: "Survey of Jazz",
    courseNumber[4]: "General Physics I"
}

course_instructors = {
    courseNumber[0]: "Z. Beasley",
    courseNumber[1]: "J. Anderson",
    courseNumber[2]: "A. Makaryus",
    courseNumber[3]: "A. Wilkins",
    courseNumber[4]: "G. Pradhan"
}

class_times = {
    courseNumber[0]: "MW 12:30pm - 1:45pm",
    courseNumber[1]: "TR 11:00am - 12:15pm",
    courseNumber[2]: "MW 9:30am - 10:45am",
    courseNumber[3]: "online asynchronous",
    courseNumber[4]: "TR 5:00pm - 6:15pm"
}

course_number = input("Enter a course number: ")
if course_number in course_names:
    print(f"Course Name: {course_names[course_number]}")
    print(f"Instructor(s): {course_instructors[course_number]}")
    print(f"Class Times: {class_times[course_number]}")
else:
    print(f"{course_number} is an invalid course number.")

#Forum task 3
days = ("Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday")
sales = []

for day in days:
    print(f"Enter the sales for {day}: ", end='')
    while True:
        try:
            salesInput = float(input())
            if salesInput < 0:
                print(f"Input was invalid. Please re-enter the sales for {day}: ",end='')
            else:
                sales.append(salesInput)
                break
        except (ValueError,TypeError,NameError):
            print(f"Input was invalid. Please re-enter the sales for {day}: ",end='')

sales.sort()
print(f"The total sales is: ${sum(sales):,.2f}")
print(f"The minimum sales this week is: ${sales[0]:,.2f}")
print(f"The maximum sales this week is: ${sales[-1]:,.2f}")

#forum task 4 - warcrimes

code = (
    ' ','--..--','.-.-.-','..--..',
    '-----','.----','..---','...--','....-','.....','-....','--...','---..','----.',
    ',-','-...','-.-.','-..','.','..-.','--.','....','..','.---','-.-','.-..','--','-.','---','.---.','--.-','.-.','...','-','..-','...-','.--','-..-','-.--','--..'
)
alphabetInput = str(input("Enter the string to be converted to Morse code\n>>")).lower()
morseCode = ''
for char in alphabetInput:
    if char == ' ':
        morseCode += code[0]
    elif char == ',':
        morseCode += code[1]
    elif char == '.':
        morseCode += code[2]
    elif char == '?':
        morseCode += code[3]
    elif char == '0':
        morseCode += code[4]
    elif char == '1':
        morseCode += code[5]
    elif char == '2':
        morseCode += code[6]
    elif char == '3':
        morseCode += code[7]
    elif char == '4':
        morseCode += code[8]
    elif char == '5':
        morseCode += code[9]
    elif char == '6':
        morseCode += code[10]
    elif char == '7':
        morseCode += code[11]
    elif char == '8':
        morseCode += code[12]
    elif char == '9':
        morseCode += code[13]
    elif char == 'a':
        morseCode += code[14]
    elif char == 'b':
        morseCode += code[15]
    elif char == 'c':
        morseCode += code[16]
    elif char == 'd':
        morseCode += code[17]
    elif char == 'e':
        morseCode += code[18]
    elif char == 'f':
        morseCode += code[19]
    elif char == 'g':
        morseCode += code[20]
    elif char == 'h':
        morseCode += code[21]
    elif char == 'i':
        morseCode += code[22]
    elif char == 'j':
        morseCode += code[23]
    elif char == 'k':
        morseCode += code[24]
    elif char == 'l':
        morseCode += code[25]
    elif char == 'm':
        morseCode += code[26]
    elif char == 'n':
        morseCode += code[27]
    elif char == 'o':
        morseCode += code[28]
    elif char == 'p':
        morseCode += code[29]
    elif char == 'q':
        morseCode += code[30]
    elif char == 'r':
        morseCode += code[31]
    elif char == 's':
        morseCode += code[32]
    elif char == 't':
        morseCode += code[33]
    elif char == 'u':
        morseCode += code[34]
    elif char == 'v':
        morseCode += code[35]
    elif char == 'w':
        morseCode += code[36]
    elif char == 'x':
        morseCode += code[37]
    elif char == 'y':
        morseCode += code[38]
    elif char == 'z':
        morseCode += code[39]

print("morse code:\n" + morseCode)



    
```
